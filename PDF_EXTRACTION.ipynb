```python
#ALL DATA EXTRACTION

import re
import PyPDF2

pdf=r"C:\Users\ANROSE\Downloads\ISO_9001_2015(en).pdf"

reader = PyPDF2.PdfReader(pdf)
num_pages = len(reader.pages)

for page in reader.pages:
    print(page.extract_text())
```

```python
#HEADING EXTRACTION

headings_sep= []       #each heading in an array inside an array
pattern = r'^\d+[\s\.].*[a-z]$'
headings=[]        #each heading as an element

for i in range(10,num_pages):
        text = re.sub(r'\t',' ', reader.pages[i].extract_text()).replace('  ',' ')
        for line in text.splitlines():
            match = re.match(pattern, line)
            if match:
              headings.append(match.group())
              headings_sep.append([match.group()])
```

```python
print(headings_sep)
```

```python
headings_separate = []

for item in headings_sep:
    inner_array = item[0].split(' ')
    number = inner_array[0]
    text = ' '.join(inner_array[1:])
    headings_separate.append([number, text])
```

```python
print(headings_separate)
```

```python
#TEXT CLEANING FROM 10TH PAGE

text_list=[]
for i in range(10,num_pages):
    line=reader.pages[i].extract_text().splitlines()
    del line[0]  #header
    del line[-1]   #footer
    text='\n'.join(line)
    text_list.append(text)
print(text_list)
```

```python
texts='\n'.join(text_list)  #string text
print(texts)
```

```python
#CLEANED TEXT

final_texts=re.sub(r'\s+', ' ',texts)
final_texts=final_texts.replace('INTERNATIONAL ST ANDARD  ISO 9001:2015(E)',' ').replace('INTERNATIONAL ST ANDARD ISO 9001:2015(E)',' ').replace('e.g.','example:').replace('NOTE In some situations, such as internet sales, a formal review is impractical for each order. Instead, the review can cover relevant product information, such as catalogues',' ').replace('.  .  ','.')#.replace('They can be conducted separately or in any combination, as is suitable for the products and services of the organization.',' ')
print(final_texts)
```

```python
print(headings)
```

```python
data=[]
for i in range(len(headings)):
    start_index = final_texts.find(headings[i]) + len(headings[i])

    if i==len(headings)-1:
        end_index=final_texts.find('Annex A')
    else:
        end_index = final_texts.find(headings[i+1])

    data.append(final_texts[start_index:end_index].strip())
```

```python
for item in data:
    print(item)
```

```python
import nltk
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk import pos_tag
```

```python
sentence_lists=[]

for i in range(3,len(data)):
    sentences = sent_tokenize(data[i])
    sentence_list=[]
    for sentence in sentences:
        sentence_list.append(sentence)
    sentence_lists.append(sentence_list)
```

```python
sentence_lists
```

```python
tagged_sentence_list = []
for sentence_list in sentence_lists:
    md_sentences = []
    for sentence in sentence_list:
        words = word_tokenize(sentence)
        tagged_words = pos_tag(words)
        for word, tag in tagged_words:
            if tag=='MD':
                if re.match(r'note\b', sentence.lower()):
                    break
                else:
                    md_sentences.append(sentence)
        if re.match(r'note\b', sentence.lower()):
                    break
    tagged_sentence_list.append(list(set(md_sentences)))

```

```python
for item in tagged_sentence_list:
    print(item)
```

```python
modified_sentences = []

for sentence_list in tagged_sentence_list:
    modified_sentence_list = []  # Store the modified sentences for each sentence_list
    for sentence in sentence_list:
        if isinstance(sentence, str):
            sentence = re.sub(r"\b\d+(\.\d+)*\b(?=\s+[a-zA-Z]+\b)", " ", sentence)
            sentence=sentence.replace("\n",'')
            split_sentence = re.split(r':\s?[a-zA-Z0-9]\)|:\s?[a-zA-Z0-9]\)|;\s?[a-zA-Z0-9]\)', sentence)
            main_sentence = split_sentence[0].strip()
            if len(split_sentence) > 1:
                subpoints = split_sentence[1:]
                subpoints = [main_sentence +' '+ subpoint.strip() for subpoint in subpoints]
                modified_sentence_list.extend(subpoints)
            else:
                modified_sentence_list.append(sentence)
        else:
            modified_sentence_list.append(sentence)
    modified_sentences.append(modified_sentence_list)
for modified_sentence_list in modified_sentences:
    print(modified_sentence_list)
```

```python
title=[]
index=[]
for i in range(3,len(headings_separate)):
    title.append(headings_separate[i][1])
    index.append(headings_separate[i][0])
```

```python
import pandas as pd
df=pd.DataFrame({'Index of title':index,'Title':title,'Data':modified_sentences})
```

```python
import pandas as pd

def generate_question(sentence):
    if pd.isnull(sentence):
        return sentence

    sentence=sentence.replace(',','').replace('.  .  ',' and ')
    sentence=re.sub(r'\.$', '', sentence)
    sentence=re.sub(r"['’]", "", sentence)
    sentence=re.sub(r"necessary.When", "necessary When", sentence)
    sentence=str(sentence)
    index = sentence.find("shall")

    if index == -1:
        return f"Is {sentence.lower()} true?"

    before_shall = sentence[:index].replace('the organization','').replace('the organizations','').strip().lower()
    after_shall = sentence[index + len("shall"):].replace('shall','').strip().lower()
    if before_shall=='when measurement traceability is a requirement or is considered by  to be an essential part of providing confidence in the validity of measurement results measuring equipment':
        return f"Is {sentence.lower()} true?"

    before_shall_words = before_shall.split()
    if len(before_shall_words) <=3:
        return f"Does {before_shall} {after_shall}?"
    else:
        if before_shall=='the scope of its quality management system' or before_shall=='the organizations quality management system' or before_shall=='documented information retained as evidence of conformity' or before_shall=='documented information of external origin determined by  to be necessary for the planning and operation of the quality management system' or before_shall=='documented information required by the quality management system and by this international standard' or before_shall=='conflicting design and development inputs' or before_shall=='the results of analysis' or before_shall=='the outputs of the management review':
            return f"Does {before_shall} {after_shall}?"

        return f"Does your organization {after_shall} {before_shall.strip()}?"

dataframe['Question'] = dataframe['Data'].apply(generate_question)
```
